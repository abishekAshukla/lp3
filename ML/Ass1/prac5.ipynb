```python
def find_out(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3 - q1
    outliers = df[((df < (q1 - (1.5 * iqr))) | (df > (q3 + (1.5 * iqr))))]
    return outliers
```

```python
outliers = find_out(df['fare_amount'])
print(len(outliers))
print(outliers.max())
print(outliers.min())
```

```python
def find_dist(longitude1, latitude1, longitude2, latitude2):
    long1, lat1, long2, lat2 = map(np.radians, [longitude1, latitude1, longitude2, latitude2])
    del_lat = lat2 - lat1
    del_long = long2 - long1
    a = np.sin(del_lat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(del_long/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) * 6371
    return c
```

```python
df['Distance'] = find_dist(
    df['pickup_longitude'],
    df['pickup_latitude'],
    df['dropoff_longitude'],
    df['dropoff_latitude']
)
```
