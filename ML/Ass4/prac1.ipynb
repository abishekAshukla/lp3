```python
currrent_x = 2
lr = 0.01
precision = 0.000001
delta_x = 1
iter_cnt = 0
maxItr = 1000
```

```python
def gradient(x):
    return 2*(x + 3)

def value(x):
    return (x+3)**2

y = []
y.append(value(currrent_x))
x = []
x.append(currrent_x)
```

```python
while precision < delta_x and iter_cnt < maxItr:
    prev_x = currrent_x
    currrent_x = prev_x -  (lr * gradient(currrent_x))
    x.append(currrent_x)
    y.append(value(currrent_x))
    delta_x = abs(prev_x - currrent_x)
    print("Iteration no: ", iter_cnt + 1, " value: ", currrent_x )
    iter_cnt += 1

print("Local minia: ", currrent_x)
```
