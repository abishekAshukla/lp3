```python
current_x = 3
rate = 0.01 #Learning rate
precision = 0.000001 #This tells us when to stop the algorithm
delta_x = 1 # previous_step_size
max_iterations = 10000
iteration_counter = 0
```

```python
#dy/dx of eqn = 2*(x+3)
def slope(x):
    return 2*(x+5)

def value_y(x):
    return (x+5)**2

y = []
x = []
y.append(value_y(current_x))
x.append(current_x)
```

```python
while precision<delta_x and iteration_counter<max_iterations:
    previous_x = current_x
    current_x = previous_x - rate * slope(previous_x)
    y.append(value_y(current_x))
    x.append(current_x)
    delta_x = abs(previous_x-current_x)
    print("Iteration ",iteration_counter+1,": ",current_x)
    iteration_counter += 1

print("Local Minima occurs at: ",current_x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_cordinate = np.linspace(-5,5,100)
plt.plot(x_cordinate,value_y(x_cordinate))
plt.scatter(x,y, color='red')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('y=(x+3)^2')
plt.show()
```
