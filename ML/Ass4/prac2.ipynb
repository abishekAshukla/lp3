```python
x_current = 2
lr = 0.01
precesion = 0.000001
delta_x = 1
no_itr = 0
max_itr = 10000
```

```python
def value(x):
    return (x + 3)**2

def gradient(x):
    return 2*(x+3)

x = []
x.append(x_current)
y = []
y.append(value(x_current))
```

```python
while precesion < delta_x and no_itr < max_itr:
    prev = x_current
    x_current = prev - (lr * gradient(prev))
    x.append(x_current)
    y.append(value(x_current))
    delta_x = abs(prev - x_current)
    print("Iteration: ", no_itr+1, " x: ", x_current)
    no_itr += 1

print("local minima of function is: ", x_current)
```
